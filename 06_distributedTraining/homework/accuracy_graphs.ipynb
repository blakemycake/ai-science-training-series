```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
n1 = pd.read_table("metrics_1node.dat", sep="\s+", header = None)
print(n1.head())
n2 = pd.read_table("metrics_2node.dat", sep="\s+", header = None)
n4 = pd.read_table("metrics_4node.dat", sep="\s+", header = None)
n8 = pd.read_table("metrics_8node.dat", sep="\s+", header = None)
```

```python
n1.columns = ['train_acc', 'train_loss', 'valid_acc', 'valid_loss', 'time_per_epochs']
n1.index.name = 'epoch'
n1.reset_index(inplace=True)
print(n1.head())

n2.columns = ['train_acc', 'train_loss', 'valid_acc', 'valid_loss', 'time_per_epochs']
n2.index.name = 'epoch'
n2.reset_index(inplace=True)
print(n2.head())

n4.columns = ['train_acc', 'train_loss', 'valid_acc', 'valid_loss', 'time_per_epochs']
n4.index.name = 'epoch'
n4.reset_index(inplace=True)
print(n4.head())

n8.columns = ['train_acc', 'train_loss', 'valid_acc', 'valid_loss', 'time_per_epochs']
n8.index.name = 'epoch'
n8.reset_index(inplace=True)
print(n8.head())
```

```python
n1.plot(x='epoch',y=['train_acc', 'valid_acc'],style=['.', '+'])
plt.savefig("accuracy_1.pdf", format="pdf", bbox_inches="tight")
plt.show()
```

```python
n2.plot(x='epoch',y=['train_acc', 'valid_acc'],style=['.', '+'])
plt.savefig("accuracy_2.pdf", format="pdf", bbox_inches="tight")
plt.show()
```
